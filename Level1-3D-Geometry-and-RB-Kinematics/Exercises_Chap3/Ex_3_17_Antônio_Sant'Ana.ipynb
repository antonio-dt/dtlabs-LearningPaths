```python
import numpy as np

# a) Find Tad and Tcd in terms of the dimensions given in the figure.

Tad = np.eye(4)
Tcd = np.eye(4)

# Frames {a} and {d} have the same orientation. Therefore:
Rad = np.array([[1,0,0],
                [0,1,0],
                [0,0,1]])

pad = np.array([[-1,1,0]]).T

# Frames {c} and {d} have diferent orientations:
# xd to yc, yd to xc and zd to -zc
Rcd = np.array([[0,1,0],
                [1,0,0],
                [0,0,-1]])

pcd = np.array([[0,0,2]]).T

Tad[:3,:4] = np.concatenate((Rad, pad), axis=1)
Tcd[:3,:4] = np.concatenate((Rcd, pcd), axis=1)

print(f'a)\nTad = {Tad}\n')
print(f'Tcd = {Tcd}\n')


# b) Find Tab given that:
Tbc = np.eye(4)
Tbc[0,3] = 4

# To find Tab
# Tab = Tad @ Tdc @ Tcb

Tdc = np.linalg.inv(Tcd)
Tcb = np.linalg.inv(Tbc)

Tab = Tad @ Tdc @ Tcb

print(f'b)\nTab = {Tab}')
```
