```python
import numpy as np
```

```python
p = np.array([1/np.sqrt(3), -1/np.sqrt(6), 1/np.sqrt(2)])

theta_x = np.deg2rad(30)
theta_y = np.deg2rad(135)
theta_z = np.deg2rad(-120)

cosx = np.cos(theta_x)
sinx = np.sin(theta_x)

cosy = np.cos(theta_y)
siny = np.sin(theta_y)

cosz = np.cos(theta_z)
sinz = np.sin(theta_z)

Rx = np.array([[1,  0,   0],
             [0, cosx, -sinx],
             [0, sinx,  cosx]])

Ry = np.array([[ cosy, 0, siny],
             [  0, 1,  0],
             [-siny, 0, cosy]])

Rz = np.array([[cosz, -sinz, 0],
             [sinz,  cosz, 0],
             [0,    0, 1]])

R = Rz @ Ry @ Rx

p_prime = R @ p

print(f"a) p'= {p_prime}\n")
print(f"b) R = ", R)
```
