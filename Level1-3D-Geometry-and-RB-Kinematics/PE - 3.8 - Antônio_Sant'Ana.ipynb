```python
import numpy as np
```

```python
def change_representation(Rab, pa):
  Rba = Rab.T
  pb = np.dot(Rba,pa)

  return pb
```

```python
Rab = np.array([[1,0,0], [0,0,-1], [0,1,0]])
pa = np.array([1,2,3])

print(change_representation(Rab,pa))
```
