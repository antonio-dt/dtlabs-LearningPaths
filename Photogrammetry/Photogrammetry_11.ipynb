# 11 - What is a Homography Matrix, how you can use that?

A homography matrix is a operator that maps points from one plane to another. It is obtained by the relation of two images of the same planar surface. It represents a projective transformation with eight degrees of freedom (estimated up to scale), requiring at least four corresponding points to calculate. It is essential for tasks such as image stitching (panoramas), rectification, and 3D pose estimation.
