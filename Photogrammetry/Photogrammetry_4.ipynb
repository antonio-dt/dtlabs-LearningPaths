## 1. Why is FLANN so fast?

FLANN (Fast Library for Approximate Nearest Neighbors) is a collection of algorithms optimized for fast nearest neighbor searches in large datasets and high-dimensional features.


The standard Brute-Force Matcher (BF) compares every descriptor in the first set against every descriptor in the second set. If you have $N$ features in both images, the complexity is $O(N^2)$. While accurate, this becomes a computational bottleneck as the number of features increases.

FLANN achieves its speed by shifting from an exact search to an approximate search. It organizes descriptors into efficient data structures:

* Randomized KD-Trees: For high-dimensional float descriptors (like SIFT), FLANN builds multiple KD-Trees. By searching across several trees simultaneously and limiting the number of leaf nodes visited, it finds the "nearest neighbor" in nearly $O(\log N)$ time.

* Hierarchical Clustering: For other types of data, it groups descriptors into clusters, allowing the search to skip entire branches of data that are mathematically distant from the query point.

The trade-off is a negligible loss in precision for a massive gain in computational efficiency.

## 2. Other Ways to Accelerate Feature Matching

### A. Binary Descriptors and Hamming Distance
Switching from float-based descriptors (SIFT) to binary-based ones (ORB, AKAZE) accelerates matching at the hardware level. Comparing two binary strings involves an XOR operation followed by a bit count, which CPUs can execute in a single clock cycle.

### B. Epipolar Geometry
If the camera's spatial relationship is partially known, we can use Epipolar Geometry. This reduces a 2D search problem into a 1D search along the Epipolar Line, significantly decreasing the number of candidates.
