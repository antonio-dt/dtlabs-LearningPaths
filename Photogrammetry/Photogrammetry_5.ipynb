# 5. In the feature point matching, mismatches will inevitably be encountered. What happens if we put the wrong match into PnP ? What methods can you think of to avoid mismatches?

PnP (Perspective-n-Point) estimates the camera's pose by mapping a set of n 3D points in the world to their corresponding 2D projections in the image.

The algorithm is built upon solving mathematical equations that minimize the reprojection error (the distance between the observed 2D point and the projected 3D point). This optimization approach makes it highly sensitive to outliers. Since a mismatch provides incorrect spatial information, the algorithm will attempt to compensate for this error during optimization. Consequently, even a few mismatches can drastically bias the results, leading to a distorted calculated pose, jittery movement, or a complete failure of the system to converge on a valid solution.

To fix that, some outlier filtering techniques can be used to enhance pnp's accuracy:
* Lowe's Ratio Test: A pre-filtering step that compares the distance of the best match to the second-best match, discarding ambiguous correspondences before they reach the geometry phase.
* Epipolar Constraint Filtering: To ensure that the matched points follow the laws of projective geometry between two views.
* RANSAC (Random Sample Consensus): To robustly estimate the pose even in the presence of many outliers.
